```python
from __future__ import annotations
import pandas as pd
import matplotlib.pyplot as plt
import subprocess
from pathlib import Path
from scipy.constants import k as kB_J_perK, eV as J_per_eV
kB_eV_per_K = kB_J_perK / J_per_eV
def repo_root() ->Path:
    try: 
        out= subprocess.check_output(["git", "rev-parse", "--show-toplevel"], stderr=subprocess.STDOUT, text=True,).strip()
        return Path(out)
    except Exception:
        here = Path.cwd().resolve()
        for p in (here, *here.parents):
            if (p / ".git").exists():
                return p
        print("could not find git repo")
        return here
ROOT = repo_root()
DATA_DIR= (ROOT / "homework-4-2")
ALT_DIR = (ROOT/ "homework -4-2")
if not DATA_DIR.exists() and ALT_DIR.exists():
    DATA_DIR = ALT_DIR
CSV_Path = DATA_DIR / "ce_thermo.csv"
print(" FOUND CSV.")


#CSV_Path = Path("/System/Volumes/Data/Users/robertmize/Desktop/ce_thermo.csv")
print("Loading:", CSV_Path)
print("Looking for:", CSV_Path.resolve())
df= pd.read_csv(CSV_Path)
df.head()
T = df["T_K"].to_numpy()
plt.figure(figsize= (7,5))
plt.plot(T, df["S_iso_eV_perK"]/ kB_eV_per_K, label= "Isolated")
plt.plot(T, df["S_soc_eV_perK"]/ kB_eV_per_K, label = "SOC only")
plt.plot(T, df["S_cfs_eV_perK"]/ kB_eV_per_K, label="SOC + CFS")
plt.xlabel("Temperature(K)")
plt.ylabel("S/k_B(dimensionless)")
plt.title("Entropy vs temperature")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig(DATA_DIR/ "S_over_k_vs_T.png", dpi =200)
plt.show()
```
